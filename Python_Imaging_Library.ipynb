#### Python Imaging Module (PIL)

```python
from PIL import Image, ImageFont, ImageFilter, ImageDraw
from matplotlib.pyplot import imshow
```

```python
file_name = 'poker.jpg'
im = Image.open(file_name)
imshow(im)
```

```python
print(im.size)
print(im.format_description)
print(im.mode)
```

#### cropping the image

```python
crop = (200,200,600,600)
cropped_image = im.crop(crop)
imshow(cropped_image)
```

#### Grayscaling an image

```python
gray_scale = im.convert('LA')
imshow(gray_scale)
```

#### Bluring an image

```python
effect = im.filter(ImageFilter.GaussianBlur(radius=20))
imshow(effect)
```
